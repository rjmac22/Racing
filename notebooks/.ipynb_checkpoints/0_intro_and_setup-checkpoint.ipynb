## 🧠 Understanding Missing Values 

We identified missing values in several key performance-related columns. Rather than treating them as errors, it's important to understand why they’re missing and when it makes sense to include them in our analysis.

### 1. `draw` – Stall Number
- Only relevant for **flat races** where horses start in stalls.
- Missing in **jumps races** like hurdles or chases.
- ✅ Use `draw` only when `type == 'Flat'`.

### 2. `or` – Official Rating
- A handicap mark assigned by the BHA.
- Missing when:
  - A horse is **unrated** (e.g., new/maiden races)
  - The race type doesn't use ratings.
- ✅ Use `or` for handicap analysis or experienced horses.
- ✅ Consider flagging `has_or = True/False`.

### 3. `ts` – TopSpeed Rating
- A Racing Post speed-based performance score.
- Missing when:
  - The horse did not finish or was too far behind.
  - The race wasn’t rated (e.g., older or lower-class races).
- ✅ Use `ts` for speed analysis, possibly alongside `rpr`.



## 💰 Understanding Missing Prize Data

While inspecting the dataset, we noticed that around **45%** of the horses had a non-null value in the `prize` column, and **100%** were missing from the `currency` column.

Rather than treating this as bad or corrupt data, it's important to understand **why** it's missing — and what that tells us.

### 🎯 Why Is `prize` Missing for Some Horses?

Not every horse in a race wins prize money. Typically:
- Only the **winner** and a few **placed runners** receive any payout.
- In small fields, prizes might go to the top 2 or 3 finishers.
- In larger or high-value races, up to 6th place (or more) might get paid.

So when `prize` is missing, it usually means the horse finished **outside the money** — i.e., did not earn a placing that awards cash.

> ✅ **Conclusion:** Missing `prize` values are not an error — they’re an expected outcome for most runners.

---

### 💱 What About `currency`?

The `currency` column is completely empty in our sample. Based on the documentation, it was meant to be parsed from the `prize_raw` string (e.g., "£3,500" → currency = "£").

In this version of the database, that parsing either hasn't happened or was deprecated in favour of normalising all prize values to a standard currency.

> ✅ **Conclusion:** `currency` can be ignored for now unless we need to reverse-engineer raw prize strings in future work.

---

### 🧠 Takeaway

Before cleaning or engineering features, always ask:
- Why is this data missing?
- Is it truly incomplete, or just *not applicable*?
- Does the absence itself carry useful meaning?

This kind of context-aware thinking avoids wasting time fixing "problems" that aren't really problems — and helps us extract richer insight from the data.



## 🐎 Different Forms of Horse Racing = Different Contexts

Just like in football, where passing plays differ from running plays, horse racing includes multiple **forms of racing** that operate under different rules and dynamics.

Our dataset includes a `type` column that distinguishes between race types:

- `Flat`: No jumps, often influenced by draw and speed
- `Hurdle`: Smaller jumps, more attrition
- `Chase`: Larger jumps, higher DNF rates
- `NH Flat` or `Bumper`: Pre-jump schooling races
- `Hunter`: Steeplechases for amateur riders

We must avoid treating all races the same. Each form has unique dynamics, risks, and performance patterns.

> 🧠 So from this point onward, we’ll analyse races by type — starting with **Flat racing**, where stalls, draw, and pace play a central role.

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style='whitegrid')  # Optional: nice default style

# Filter for flat races only
flat_df = df[df['type'].str.lower() == 'flat'].copy()

# Recalculate win column (just to be safe)
flat_df['won'] = (flat_df['pos'] == 1).astype(int)

# Win rate in flat races
flat_win_rate = flat_df['won'].mean()
print(f"Win rate in Flat races: {flat_win_rate:.2%}")

# Bar chart for Flat races
sns.countplot(x='won', data=flat_df)
plt.title('Win Distribution in Flat Races')
plt.xticks([0, 1], ['Did Not Win', 'Won'])
plt.xlabel('')
plt.ylabel('Number of Horses')
plt.show()
```
